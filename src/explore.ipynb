```python
import requests
from bs4 import BeautifulSoup
import re 
import pandas as pd

# Seleccionar el recurso a descargar
resource_url = "https://es.wikipedia.org/wiki/Leucocito"

# Petición para descargar el fichero de Internet
response = requests.get(resource_url)

# Si la petición se ha ejecutado correctamente (código 200), entonces el fichero se ha podido descargar

soup = BeautifulSoup(response.text, 'html.parser')
soup   # Corregir 'html' a 'html.parser'
    
```

```python
tables = soup.find_all("table")
tables
```

```python
leucocito = soup.find_all("table", class_="wikitable")
leucocito
```

```python
import pandas as pd 
# Obtener las filas de la tabla
rows = leucocito[0].find_all("tr")

# Inicializar listas para almacenar los datos de las filas y columnas
data = []
columns = []

# Obtener los encabezados de la tabla (nombres de columnas)
for th in rows[0].find_all("th"):
    columns.append(th.text.strip())

# Obtener los datos de las filas
for row in rows[1:]:
    row_data = [td.text.strip() for td in row.find_all("td")]
    data.append(row_data)

# Crear el DataFrame
df = pd.DataFrame(data)
# Crear el DataFrame
df = pd.DataFrame(data, columns=columns)

df
```

```python
import pandas as pd

df['Porcentaje aproximado en adultos'] = df['Porcentaje aproximado en adultos'].astype(str).str.replace('[\%,]', '').astype(float)
df['Diámetro (μm)'] = df['Diámetro (μm)'].astype(str).str.extract('(\d+)').astype(float)

df['Ultimo DataFrame'] = df['Porcentaje aproximado en adultos'] / df['Diámetro (μm)']

df
```
